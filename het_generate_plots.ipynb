```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import re
import matplotlib.ticker
```

```python
def extract_param(fname, param):
    # Define the regular expression pattern to match the parameter and its value
    pattern = r"{}(\d+)".format(param)
    
    # Search for the pattern in the file name
    match = re.search(pattern, fname)
    
    if match:
        # Extract and return the matched value as an integer
        return int(match.group(1))
    else:
        # Return None if the parameter is not found in the file name
        return None
```

# Experiment 1 : Length of trial

```python
f_list = [
    "./data/eval_student_het_test_l75_T3_j5_n10000.csv",
    "./data/eval_student_het_test_l100_T3_j5_n10000.csv",
    "./data/eval_student_het_test_l125_T3_j5_n10000.csv",
    "./data/eval_student_het_test_l150_T3_j5_n10000.csv",
    # "./data/eval_student_het_test_l200_T3_j5_n10000.csv"
    ]
nn_mse = []
srcs_mse = []
bb_aware_mse = []
for fname in f_list:
    eval = pd.read_csv(fname)
    nn_mean = eval.NN_mean.mean()
    nn_mse.append(nn_mean)
    srcs_mean = eval.SRCs_mean.mean()
    srcs_mse.append(srcs_mean)
    bb_aware_mean = eval.BB_Aware_mean.mean()
    bb_aware_mse.append(bb_aware_mean)
    print(f"NN Mean = {nn_mean:.3e}")
    print(f"SRCs Mean = {srcs_mean:.3e}")
    print(f"BB-Aware Mean = {bb_aware_mean:.3e}")
time_vec = [extract_param(fname, param="l")*(extract_param(fname, param="T")-1) for fname in f_list]
plt.plot(time_vec, nn_mse, "-ob",label="NN MSE")
plt.plot(time_vec, srcs_mse, "--sr",label="T-SRCs MSE")
plt.plot(time_vec, bb_aware_mse, "--^c",label="T-BB-Aware MSE")
plt.xlabel("Number of Time Slots")
plt.yscale('log')
ax=plt.gca()
plt.ylim(0.8e-3, 2e-2)
ax.set_yticks([1e-3, 5e-3, 1e-2])
ax.get_yaxis().set_major_formatter(matplotlib.ticker.LogFormatterSciNotation())
# ax.get_yaxis().get_major_formatter().labelOnlyBase = False
plt.ylabel("Average MSE (normalized)")
plt.legend()
plt.grid(which="both")
plt.title(f"Performance Against Number of Time Slots")
plt.savefig(f"./plots/het_exp1_draft.png")
```
