```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Experiment 1 : Length of trial

```python
f_list = ["./data/eval_student_het_test_l75_T3_j5_n10000.csv","./data/eval_student_het_test_l100_T3_j5_n10000.csv","./data/eval_student_het_test_l125_T3_j5_n10000.csv", "./data/eval_student_het_test_l150_T3_j5_n10000.csv", "./data/eval_student_het_test_l200_T3_j5_n10000.csv"]
nn_mse = []
srcs_mse = []
for fname in f_list:
    eval = pd.read_csv(fname)
    nn_mean = eval.NN_mean.mean()
    nn_mse.append(nn_mean)
    srcs_mean = eval.SRCs_mean.mean()
    srcs_mse.append(srcs_mean)
    print(f"NN Mean = {nn_mean:.3e}")
    print(f"SRCs Mean = {srcs_mean:.3e}")
length_of_trial_vec = [75, 100, 125, 150, 200]
plt.plot(length_of_trial_vec, nn_mse, "-ob",label="NN MSE")
plt.plot(length_of_trial_vec, srcs_mse, "--sr",label="T-SRCs MSE")
plt.xlabel("Length of trial (timeslots)")
plt.ylabel("Average MSE (normalized)")
plt.legend()
plt.grid()
plt.title(f"Varying the length of the BB trial in T-SRCs")
plt.savefig(f"./plots/het_exp1_new_trial.png")
```
