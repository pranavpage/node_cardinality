```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

## Experiment 1 


```python
f_list = ["./data/eval_student_reform-end-to-end_l50_j5_n10000.csv", "./data/eval_student_exp1_l100_j5_n10000.csv", "./data/eval_student_exp1_l150_j5_n10000.csv", "./data/eval_student_exp1_l200_j5_n10000.csv"]
nn_mse = []
bnb_mse = []
for fname in f_list:
    eval = pd.read_csv(fname)
    nn_mean = eval.NN_mean.mean()
    nn_mse.append(nn_mean)
    bnb_mean = eval.BnB_mean.mean()
    bnb_mse.append(bnb_mean)
    print(f"NN Mean = {nn_mean:.3e}")
    print(f"Bnb Mean = {bnb_mean:.3e}")
length_of_trial_vec = [50, 100, 150, 200]
plt.plot(length_of_trial_vec, nn_mse, "-ob",label="NN MSE")
plt.plot(length_of_trial_vec, bnb_mse, "--sr",label="BnB MSE")
plt.xlabel("Length of trial (timeslots)")
plt.ylabel("Average MSE (normalized)")
plt.legend()
plt.grid()
plt.title(f"Varying the length of the BnB trial")
plt.savefig(f"./plots/exp1.png")
```

## Experiment 2

```python
f_list = ["./data/eval_student_reform-end-to-end_l50_j5_n10000.csv", "./data/eval_student_exp1_l50_j10_n10000.csv", "./data/eval_student_exp1_l50_j20_n10000.csv", "./data/eval_student_exp1_l50_j50_n10000.csv"]
nn_mse = []
bnb_mse = []
for fname in f_list:
    eval = pd.read_csv(fname)
    nn_mean = eval.NN_mean.mean()
    nn_mse.append(nn_mean)
    bnb_mean = eval.BnB_mean.mean()
    bnb_mse.append(bnb_mean)
    print(f"NN Mean = {nn_mean:.3e}")
    print(f"Bnb Mean = {bnb_mean:.3e}")
jumps_vec = [5, 10, 20, 50]
bnb_mse=np.array(bnb_mse)
nn_mse=np.array(nn_mse)
plt.plot(jumps_vec, nn_mse, "-ob",label="NN MSE")
plt.plot(jumps_vec, bnb_mse, "--sr",label="BnB MSE")
# plt.plot(jumps_vec, bnb_mse-nn_mse, "-ok")
plt.xlabel("jumps")
plt.ylabel("Average MSE")
plt.title(f"Varying the number of jumps in the TPM")
plt.legend()
plt.grid()
plt.savefig(f"./plots/exp2.png")
```
