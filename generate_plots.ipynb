```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

## Experiment 1 


```python
f_list = ["./data/eval_student_reform-end-to-end_l50_j5_n10000.csv", "./data/eval_student_exp1_l100_j5_n10000.csv", "./data/eval_student_exp1_l150_j5_n10000.csv", "./data/eval_student_exp1_l200_j5_n10000.csv"]
nn_mse = []
bnb_mse = []
for fname in f_list:
    eval = pd.read_csv(fname)
    nn_mean = eval.NN_mean.mean()
    nn_mse.append(nn_mean)
    bnb_mean = eval.BnB_mean.mean()
    bnb_mse.append(bnb_mean)
    print(f"NN Mean = {nn_mean:.3e}")
    print(f"Bnb Mean = {bnb_mean:.3e}")
length_of_trial_vec = [50, 100, 150, 200]
plt.plot(length_of_trial_vec, nn_mse, "-ob",label="NN MSE")
plt.plot(length_of_trial_vec, bnb_mse, "--sr",label="SRCs MSE")
plt.xlabel("Length of trial (timeslots)")
plt.ylabel("Average MSE (normalized)")
plt.legend()
plt.grid()
plt.title(f"Varying the length of the BnB trial")
plt.savefig(f"./plots/exp1.png")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

f_list = [
    "./data/eval_student_homo_reform_test_l75_j5_n10000.csv",
    "./data/eval_student_homo_reform_test_l100_j5_n10000.csv",
    "./data/eval_student_homo_reform_test_l150_j5_n10000.csv",
    "./data/eval_student_homo_reform_test_l200_j5_n10000.csv",
    "./data/eval_student_homo_reform_test_l300_j5_n10000.csv"
]

nn_mse = []
srcs_mse = []
bb_aware_mse = []

nn_std = []
srcs_std = []
bb_aware_std = []

for fname in f_list:
    eval = pd.read_csv(fname)

    nn_mean = eval.NN_mean.mean()
    nn_mse.append(nn_mean)

    nn_std_ = eval.NN_mean.std()
    nn_std.append(nn_std_)

    srcs_mean = eval.SRCs_mean.mean()
    srcs_mse.append(srcs_mean)

    srcs_std_ = eval.SRCs_mean.std()
    srcs_std.append(srcs_std_)

    bb_aware_mean = eval.BB_aware_mean.mean()
    bb_aware_mse.append(bb_aware_mean)

    bb_aware_std_ = eval.BB_aware_mean.std()
    bb_aware_std.append(bb_aware_std_)

    print(f"NN Mean = {nn_mean:.3e}, SRCs Mean = {srcs_mean:.3e}, BB-Aware Mean = {bb_aware_mean:.3e}")

length_of_trial_vec = [75, 100, 150, 200, 300]
plt.errorbar(length_of_trial_vec, nn_mse, 2 * np.array(nn_std), fmt="-ob", capsize=3, label="NN MSE (2-sigma)")
plt.errorbar(length_of_trial_vec, srcs_mse, 2 * np.array(srcs_std), fmt="--sr", capsize=3, label="SRCs MSE (2-sigma)", alpha=0.7)
plt.errorbar(length_of_trial_vec, bb_aware_mse, 2 * np.array(bb_aware_std), fmt="--^c", capsize=3, label="BB-Aware MSE (2-sigma)", alpha=0.7)

plt.xlabel("Length of trial (timeslots)")
plt.ylabel("Average MSE (normalized)")
plt.legend()
plt.grid()
plt.title(f"Varying the length of the BnB trial")

# Set y-axis ticks to scientific notation
plt.ticklabel_format(axis="y", style="sci", scilimits=(0, 0))

# Show the plot
plt.show()
```
